A magician is so good at shuffling that his shuffles are always perfect in-shuffles. That means his shuffles are done in the following two steps:

1. Split the pile of cards exactly in half.
2. Interleave the top half with the bottom half such that every second card is from the top half.

If he started with six cards and shuffled thrice, he would get the original arrangement back.

![title](https://ds055uzetaobb.cloudfront.net/image_optimizer/5d36dbe8c7a4240ed93d1a201a71d3920ebc285e.png)

If he started with a deck of 52 cards, how many shuffles would he need to get the original arrangement back?

```python
cards = range(1,53)

def shuffle(cards, shuffle_count = 1):
    #know the original order
    print('The cards are:\n' + str(cards) + '\nAfter shuffle #' + str(shuffle_count-1)+'\n')
    orig = list(range(1, len(cards)+1))
    
    #find the half length
    half = len(cards)//2
    
    #split the deck
    first_half = [x for x in cards[:half]]
    second_half = [x for x in cards[half:]]
    
    #reorder the cards
    cards = []
    for i in range(half):
        cards.append(second_half[i])
        cards.append(first_half[i])
       
    if cards == orig:
        return 'The cards are back in their original order after shuffle #' + str(shuffle_count) + '!'
    else:
        return shuffle(cards, shuffle_count+1)

print(shuffle(list(range(1,7))))
```

# Removing Every Other

Have integers 1 through 12345 (inclusive).
+ Step1: Remove all numbers in odd positions (1, 3, 5, 7, ..., 12345)
+ Step2: Take whats left and remove all numbers in even positions (4, 8, 12, 16 ...)
+ Step3: Continue until only one number left

```python
x = range(12345)
# [item for item in x]

def remove(x):
    evens = []
    for i in range(len(x)):
        if i%2==1:
            evens.append(x[i])
#             print(evens)
    odds = []
    for i in range(len(evens)):
        if i%2==0:
            odds.append(evens[i])
#             print(odds)
    if len(odds)<2:
        return odds[0]+1
    else:
        return remove(odds)
            
remove(x)
            
```
